```python

means = df.groupby('Species').mean()
sepal_length_means = means['SepalLengthCm']
sepal_width_means = means['SepalWidthCm']
petal_length_means = means['PetalLengthCm']
petal_width_means = means['PetalWidthCm']
species = means.index

# Define width of each bar
bar_width = 0.2

# Set position of bar on X axis
r1 = range(len(species))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Plotting the bars
plt.bar(r1, sepal_length_means, color='b', width=bar_width, edgecolor='grey', label='Sepal Length')
plt.bar(r2, sepal_width_means, color='g', width=bar_width, edgecolor='grey', label='Sepal Width')
plt.bar(r3, petal_length_means, color='r', width=bar_width, edgecolor='grey', label='Petal Length')
plt.bar(r4, petal_width_means, color='y', width=bar_width, edgecolor='grey', label='Petal Width')

# Adding labels
plt.xlabel('Species', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(species))], species)
plt.ylabel('Mean Values')
plt.title('Mean Sepal and Petal Measurements by Species')

plt.legend()
plt.show()
```

```python
label_name=["Iris-setosa","Iris-versicolor","Iris-virginica"]
```

```python
df["Species"].replace({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2},inplace=True)
```

```python
X=df.drop(columns="Species",axis=1)
Y=df["Species"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
dt = DecisionTreeClassifier(max_depth=5,random_state=42,criterion='gini', splitter='best')
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
```

```python

from sklearn.metrics import f1_score, recall_score, precision_score, roc_auc_score, confusion_matrix, classification_report

# Calculate F1 score
f1 = f1_score(y_test, y_pred, average='weighted')

# Calculate recall
recall = recall_score(y_test, y_pred, average='weighted')

# Calculate precision
precision = precision_score(y_test, y_pred, average='weighted')

pred1=dt.predict_proba(X_test)
roc_auc = roc_auc_score(y_test,pred1, multi_class='ovr', average='weighted')






print("F1 Score:", f1)
print("Recall:", recall)
print("Precision:", precision)
print("ROC AUC Score:", roc_auc)


```

```python

print("Classification Report:")
print(classification_report(y_test, y_pred,target_names=label_name))
```

```python

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True,cmap="Accent",xticklabels=label_name,yticklabels=label_name)
plt.show()
```
